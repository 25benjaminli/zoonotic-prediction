```python
import gzip
import os
from Bio import SeqIO, Entrez
import pandas as pd
```

```python
#unzip each gz file in the refseq directory
num = 0
for folder in os.listdir('refseq/viral'):
    if folder != '.DS_Store':
        for file in os.listdir(f'refseq/viral/{folder}'):
            if file.endswith('.gz'):
                with gzip.open(f'refseq/viral/{folder}/{file}', 'rb') as f_in:
                    with open(f'seqs/{file[:-3]}', 'wb') as f_out:
                        f_out.write(f_in.read())
                        num+=1

print(f'Unzipped {num} files')
```

```python
families = ['Adenoviridae','Papillomaviridae',
'Herpesviridae','Baculoviridae','Poxviridae','Parvoviridae',
'Retroviridae','Reoviridae','Picornaviridae',
'Flaviviridae','Caliciviridae','Coronaviridae',
'Bunyaviridae','Orthomyxoviridae','Paramyxoviridae', 'Anelloviridae']

df = pd.DataFrame(columns=['family','accession', 'name','DNA Sequence'])
print(len(os.listdir('seqs')))
for file in os.listdir('seqs'):
    print(file)
    n = 0


    # print(type(i))
    fasta_sequences = SeqIO.parse(open(f"seqs/{file}"),'genbank')
    # for record in fasta_sequences:
    #     print(record.id)
    #     print(record.seq)
    
    fastas = [x for x in fasta_sequences]
    # print(fastas)
    for fasta in fastas:
        # print(vars(fasta))
        accession, sequence, taxonomy, name = fasta.id, str(fasta.seq), fasta.annotations['taxonomy'], fasta.description
        family = None
        # print(taxonomy)
        for tax in taxonomy:
            if tax in families:
                family = tax
                break
        if family == None:
            family = 'Other'

                
        df = pd.concat([pd.DataFrame({'family':family,'accession':accession, 'name':name.split(",")[0],'DNA Sequence':sequence}, index=[0]),df], ignore_index=True)
        
        # print(name)
        # print(sequence)
        n+=1
    print(n)
    
```
